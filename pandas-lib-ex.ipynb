```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(pd.__version__)
```

```python
mydataset = {
    "cars":["BMW","Volvo","Ford"],
    "passings":[3, 7, 2]
}
myvar = pd.DataFrame(mydataset)
print(myvar)
```

```python
# Pandas Series

a = [1, 5, 7, 9]
myvar = pd.Series(a)
print(myvar)

# get specific value from series
print(myvar[0])
```

```python
# Label for our text instead of index
a = [5, 7, 8]
myvar = pd.Series(a, index = ["X", "Y", "Z"])

print(myvar)
print(myvar["Y"])
```

```python
# Key value objects as series

calaries = {"day1": 430, "day2": 450, "day3": 500}
myvar = pd.Series(calaries)
print(myvar)
```

```python
# Key value objects as series include only few items from dictionary

users = {"User1": 30, "User2": 40, "User3": 50}
myvar = pd.Series(users, index=["User1","User3"])
print(myvar)
```

```python
## DATAFRAMEs - datasets in pandas like multi-dimensional table called dataframes
# Series - like column , dataframe- like whole table

data ={
    "calaries": [450, 345, 500],
    "duration": [5, 10, 12]
}

# load data into dataframe object

df = pd.DataFrame(data, index=["d1","d2","d3"])
print(df)

# return the specific row using 'loc'
print(df.loc["d1"])
```
