## 6. Dividing Integers

```python
# 1. True Division(/) - It returns with decimal
print(10/7)

# 2. Floor Division(//) - It returns only quotient without decimal
print(10//7)

# 3. Modulus Division(%) - It returns remaining reminder values
print(10%7)
```

## 13. Logical Operators

```python
# and operator
password = "password"
confirm_password = "password"
if password and confirm_password:
    print("Login Success!")
else:
    print("Logic Failed!")

# or operator

is_premium_member = False
cart_total = 150.00

if is_premium_member or cart_total >= 100:
    discount = 1.5
    print("Discount Applied!")
    
else:
    print("Discount not applied!")

# not operator

is_banned = False
if not is_banned:
    print("Accessed")
else:
    print("Banned")
```
